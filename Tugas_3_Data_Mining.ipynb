```python
outlier_data = {
    'ratings_count' : [4780653, 10439],
    'work_ratings_count' : [4942365, 12970],
    'ratings_1, ratings_2, ratings_3, ratings_4, ratings_5' : [3011543, 203],
    'original_publication_year' : [2015.0, 1997.0]
}

outlier_df =pd.DataFrame(outlier_data)

df = pd.concat([df, outlier_df], ignore_index=True)
df
```

```python
# Hitung kuartil pertama (Q1)
q1 = df['books_count'].quantile(0.25)
print("Q1 : ", q1)
# Hitung kuartil ketiga (Q3)
q3 = df['books_count'].quantile(0.75)
print("Q3 : ", q3)
iqr = q3 - q1
# Hitung batas bawah (Lower Bound) dan batas atas (Upper Bound) untuk outlier
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Temukan outlier dalam DataFrame
outliers = df[(df['books_count'] < lower_bound) | (df['books_count'] > upper_bound)]

print("Outliers:")
print(outliers)
```

```python
data_books_count = df["books_count"]
# Buat histogram
n, bins, patches = plt.hist(data_books_count, bins=5, color='blue', edgecolor='black')

# Tambahkan Label
plt.title('Histogram Nilai')
plt.xlabel('books_count')
plt.ylabel('Frekuensi')

# Tampilkan rentang frekuensi di sumbu x
bin_centers = 0.5 * (bins[:-1] + bins[1:])
plt.xticks(bin_centers, ['{:.0f}-{:.0f} ' .format(bins[i], bins[i+1]) for i in range(len(bins)-1)])

# Tampilkan histogram
plt.show()
```

```python
# Mengidentifikasi baris outlier berdasarkan ciri-ciri yang unik

outliers_condition = ((df['books_count'] > upper_bound) | (df['books_count'] < lower_bound)) | (df['ratings_count'] < 30) | (df['ratings_count'] > 200)

# Drop baris yang memenuhi kondisi di atas (outliers)
df_cleaned = df[~outliers_condition]  # Tanda ~ untuk negasi (tidak termasuk kondisi outlier)

# Menampilkan data setelah penghapusan outliers
print(df_cleaned.tail())
```
