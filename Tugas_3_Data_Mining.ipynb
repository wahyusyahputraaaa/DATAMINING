```python
import pandas as pd
import numpy as np
# Importing and Exporting Data
df = pd.read_csv('/content/drive/MyDrive/Dataset/hotel_bookings.csv')
df
```

```python
# Melihat Datatype dan Spesifik pada Dataset
df.info()
```

```python
# Mengecek apakah ada Missing Values
df.isnull().sum()
```

```python
# Dropping unnnecessart columns atau
# hapus data yang diperlukan karena keseluruhan data memiliki nilai kosong
columns_to_drop = ['children', 'country', 'agent', 'company']
df.drop(columns=columns_to_drop, inplace=True)
df
```

```python
# Menjadikan identifier sebagai index agar memudahkan mengambil sebuah data
df.set_index('hotel', inplace=True)
df
```

```python
# Menghitung nilai max, min, dan range untuk 'Life expectancy'
arrival_date_week_number_max = df['arrival_date_week_number'].max()
arrival_date_week_number_min = df['arrival_date_week_number'].min()
arrival_date_week_number_range = arrival_date_week_number_max - arrival_date_week_number_min

# Menampilkan hasil
print(f"Max arrival_date_week_number: {arrival_date_week_number_max}")
print(f"Min arrival_date_week_number: {arrival_date_week_number_min}")
print(f"Range arrival_date_week_number: {arrival_date_week_number_range}")
```

```python
mean = sum(df['arrival_date_week_number']) / df['arrival_date_week_number'].count()
mean
```

```python
median = df['arrival_date_week_number'].median()
median
```

```python
print(df['arrival_date_week_number'].mode())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data_arrival_date_week_number = df["arrival_date_week_number"]
# Buat histogram
n, bins, patches = plt.hist(data_arrival_date_week_number, bins=5, color='blue', edgecolor='black')

# Tambahkan Label
plt.title('Histogram Distribusi Minggu Ke Depan')
plt.xlabel('arrival_date_week_number')
plt.ylabel('Frekuensi')

# Tampilkan rentang frekuensi di sumbu x
bin_centers = 0.5 * (bins[:-1] + bins[1:])
plt.xticks(bin_centers, ['{:.0f}-{:.0f} ' .format(bins[i], bins[i+1]) for i in range(len(bins)-1)])

# Tampilkan histogram
plt.show()
```
