```python
import pandas as pd
import numpy as np
# Importing and Exporting Data
df = pd.read_csv('/content/drive/MyDrive/Dataset/BL-Flickr-Images-Book.csv')
df
```

```python
# Melihat Datatype dan Spesifik pada Dataset
df.info()
```

```python
# Mengecek apakah ada Missing Values
df.isnull().sum()
```

```python
# Dropping unnnecessart columns atau
# hapus data yang diperlukan karena keseluruhan data memiliki nilai kosong
columns_to_drop = ['Edition Statement', 'Corporate Author', 'Corporate Contributors', 'Former owner', 'Engraver']
df.drop(columns=columns_to_drop, inplace=True)
df
```

```python
# Menjadikan identifier sebagai index agar memudahkan mengambil sebuah data
df.set_index('Identifier', inplace=True)
df
```

```python
# Renaming columns
df.rename(columns={
    'Place of Publication': 'Tempat_penerbitan',
    'Date of Publication': 'Tahun_penerbit',
    'Publisher': 'Penerbit',
    'Title': 'Judul',
    'Author': 'Penulis',
    'Contributors': 'Kontributor',
    'Issuance type': 'Jenis_penerbitan',
    'Flickr URL': 'URL_Flickr',
    'Shelfmarks': 'Penanda_Rak'

}, inplace=True)
# Tampilan DataFrame setelah perubahan nama kolom
df
```
