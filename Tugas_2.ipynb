```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```python
data = pd.read_csv('/content/drive/MyDrive/500_Person_Gender_Height_Weight_Index.csv')

```

```python
print(data.head())
```

```python
# Kelompokkan data berdasarkan gender
grouped = data.groupby('Gender')
# Hitung statistik deskriptif
print(grouped['Height'].describe())
print(grouped['Weight'].describe())
data['Gender'] = data['Gender'].map({'Female': 0, 'Male': 1})
plt.figure(figsize=(10, 6))
plt.boxplot([data[data['Gender'] == 0]['Index'], data[data['Gender'] == 1]['Index']], labels=['Female', 'Male'])
plt.title('Distribusi Indeks BMI Berdasarkan Gender')
plt.ylabel('Indeks BMI')
plt.show()





```
